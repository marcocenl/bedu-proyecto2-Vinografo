# Forma Novata

```python
def calidad(columna):
  if columna <= 6:
    return 'low'
  elif columna <= 8:
    return 'medium'
  else:
    return 'high'
```

```python
# apply <- particularmente lento
df['quality_categorical'] = df['quality'].apply(calidad)
```

# Forma Vectorizado
